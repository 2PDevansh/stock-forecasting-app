#### Ticker Objects

```python
import yfinance as yf

dat = yf.Ticker("MSFT")    #MSFT='Microsoft Corp'

# get historical market data
dat.history(period='1mo')   #1 month data

# options
dat.option_chain(dat.options[0]).calls # first expiration

# get financials
dat.balance_sheet          # balance sheet
dat.cashflow              # cash flow
dat.quarterly_income_stmt   # quarterly income statement

# dates
dat.calendar

# general info
dat.info         # general information

# analysis
dat.analyst_price_targets      # analyst price targets

# websocket
 dat.live()             # get real-time market data
```

### Multiple Ticker Objects

```python
import yfinance as yf

tickers = yf.Tickers('msft aapl goog')

# access each ticker using (example)
tickers.tickers['MSFT'].info                  # Microsoft Corp
tickers.tickers['AAPL'].history(period="1mo") # Apple Inc
tickers.tickers['GOOG'].actions               # Google Inc

# websocket
tickers.live()
```
