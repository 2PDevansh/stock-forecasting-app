```python
from alpha_vantage.timeseries import TimeSeries
import matplotlib.pyplot as plt
ts = TimeSeries(key='8RM95GGIN3EY84ND', output_format='pandas', indexing_type='date')
data, meta_data = ts.get_intraday(symbol='GOOGL', interval='60min', outputsize='full')
# Plotting the closing prices
data['4. close'].plot()
plt.title('Intraday Times Series for the MSFT stock (60 min)')
plt.show()
```

```python
# Prepare a DataFrame with the required feature columns (use the fetched 'data' DataFrame)
ts = data.copy()
# Rename the Alpha Vantage column names to simpler names
ts.rename(columns={
	'1. open': 'Open',
	'2. high': 'High',
	'3. low': 'Low',
	'4. close': 'Close',
	'5. volume': 'Volume'
}, inplace=True)

# Create technical features used by the model
ts['MA50'] = ts['Close'].rolling(window=50).mean()
ts['MA200'] = ts['Close'].rolling(window=200).mean()
# Volatility as rolling std of returns (50-period)
ts['Volatility'] = ts['Close'].pct_change().rolling(window=50).std()

# Drop rows with NaNs introduced by rolling windows
ts.dropna(inplace=True)

# Now 'ts' contains the columns ['Close','MA50','MA200','Volatility'] required by downstream cells.
# Scaling will be performed in the subsequent cell (do not redefine scaler here to avoid duplication).
```

```python
# One day ahead Stock Price Prediction using LSTM for GOOGL
import numpy as np
from sklearn.preprocessing import MinMaxScaler
from keras.models import Sequential
from keras.layers import LSTM, Dense, Dropout
from sklearn.metrics import mean_squared_error
import tensorflow as tf
import math
from sklearn.model_selection import train_test_split
import pickle

# Scale features
scaler_g = MinMaxScaler(feature_range=(0, 1))
scaled_data_g = scaler_g.fit_transform(ts[['Close', 'MA50', 'MA200', 'Volatility']])
# Save the scaler for future use
pickle.dump(scaler_g, open("g_scaler.pkl", "wb"))
pickle.dump(scaled_data_g, open("g_scaled_data.pkl", "wb"))

# Create sequences
def create_sequences(data, seq_length):
    X, y = [], []
    for i in range(len(data) - seq_length):
        X.append(data[i:i + seq_length])
        y.append(data[i + seq_length, 0])  # predict Close
    return np.array(X), np.array(y)

seq_length = 60
X, y = create_sequences(scaled_data_g, seq_length)

# Split the data into training and testing sets
# For time series keep chronological order (no shuffle)
x_train, x_test, y_train, y_test = train_test_split(X, y, test_size=0.2, shuffle=False)
# Build LSTM model
model = Sequential()
model.add(LSTM(units=100, return_sequences=True, input_shape=(X.shape[1], X.shape[2]))) # First LSTM layer
model.add(Dropout(0.2)) 
model.add(LSTM(units=50, return_sequences=False)) # Last LSTM layer
model.add(Dropout(0.2)) # Dropout layer to prevent overfitting
model.add(Dense(units=1))  # Predicting the closing price
model.compile(optimizer='adam', loss='mean_squared_error') # Compile the model
# Train the model
model.fit(x_train, y_train, epochs=150, batch_size=32, validation_data=(x_test, y_test))
# Make predictions
predictions = model.predict(x_test)
#add 3 columns of zeros (dummy placeholders):
predictions = scaler_g.inverse_transform(np.concatenate((predictions, np.zeros((predictions.shape[0], 3))), axis=1))[:,0] # Rescale back to original values
# Rescale y_test to original values for accurate RMSE calculation
y_test_rescaled = scaler_g.inverse_transform(np.concatenate((y_test.reshape(-1, 1), np.zeros((y_test.shape[0], 3))), axis=1))[:,0]
# Evaluate the model
rmse = math.sqrt(mean_squared_error(y_test_rescaled, predictions))
print(f'Root Mean Squared Error: {rmse}')
model.save('g_model.h5')
# Stock Price Prediction using LSTM
model = tf.keras.models.load_model('g_model.h5')
# Prepare the latest data for prediction
latest_data = scaled_data_g[-seq_length:]  # Last 'seq_length' days
latest_data = np.expand_dims(latest_data, axis=0)  # Reshape for model
# Predict the next day's closing price
predicted_price = model.predict(latest_data)
# Rescale back to original values
predicted_price = scaler_g.inverse_transform(np.concatenate((predicted_price, np.zeros((predicted_price.shape[0], 3))), axis=1))[:, 0] 
print(f'Predicted Next Day Closing Price: {predicted_price[0]}')
```
